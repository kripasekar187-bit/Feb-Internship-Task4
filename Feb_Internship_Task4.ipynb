```python
# Problem 1: Mobile Recharge Validation System
#This program validates a mobile recharge request using function and while loop.

def validate_recharge():
    # List of valid recharge plans
    valid_plans = [199, 299, 399, 599]
    # Retry loop
    while True:
        amount = int(input("Enter recharge amount: ₹"))
        # Check minimum amount
        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.")
            continue
        plan = int(input("Enter plan (199, 299, 399, 599): "))
        # Validate plan
        if plan not in valid_plans:
            print("Invalid plan selected. Try again.")
            continue
        print("Recharge Successful!")
        break
# Function call
validate_recharge()

```

```python
# Problem 2: Inventory Reorder Alert System
#This program checks product stock and prints reorder alerts.

# Sample inventory data
products = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Wheat": 30
}
def check_inventory(stock_data):
    # Loop through dictionary
    for product, quantity in stock_data.items():
        # Check stock condition
        if quantity < 15:
            print(product, "- Reorder Alert")
        else:
            print(product, "- Stock OK")
# Function call
check_inventory(products)

```

```python
# Problem 3: Student Result Processing System
#This program calculates average marks and determines pass or fail.

# Sample marks list
marks = [60, 45, 70, 55, 40]
def process_result(marks_list):
    # Initialize total
    total = 0
    # Calculate total using loop
    for mark in marks_list:
        total += mark
    # Calculate average
    average = total / len(marks_list)
    # Determine pass or fail
    if average >= 50:
        return "Pass"
    else:
        return "Fail"
# Function call
result = process_result(marks)
print("Result:", result)

```

```python
# Problem 4: Cab Fare Estimator with Retry Option
#This program calculates cab fare based on distance and peak hour charges.

def calculate_fare():
    while True:
        # Get distance from user
        km = float(input("Enter distance in km: "))
        # Calculate base fare
        fare = 50 + (12 * km)
        # Check peak hour
        peak = input("Is it peak hour? (yes/no): ").lower()
        if peak == "yes":
            fare += fare * 0.25
        # Display fare
        print("Total Fare: ₹", round(fare, 2))
        # Retry option
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Estimator!")
            break
# Function call
calculate_fare()

```

```python
# Problem 5: Employee Attendance Eligibility Checker
#This program calculates attendance percentage and checks eligibility.

# Sample attendance list
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
def check_eligibility(attendance_list):
    # Initialize present count
    present_count = 0
    # Count present days
    for status in attendance_list:
        if status == "P":
            present_count += 1
    # Calculate attendance percentage
    percentage = (present_count / len(attendance_list)) * 100
    # Check eligibility condition
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
# Function call
result = check_eligibility(attendance)
print("Attendance Status:", result)

```

```python
# Problem 6: Password Strength Checker
#This program checks whether a password is strong based on given rules.

def check_password_strength(password):
    has_digit = False
    has_special = False
    # Check minimum length
    if len(password) < 8:
        return "Weak Password"
    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    # Final strength check
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
# User input
password = input("Enter your password: ")
result = check_password_strength(password)
print(result)

```
